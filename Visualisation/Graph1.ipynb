```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df= sns.get_dataset_names()
```

```python
df= sns.load_dataset('titanic')
```

```python
df
```

```python
#countplot: Survival based on gender
sns.countplot(df,x='survived', hue='sex', palette="bright")
plt.title('Survived vs Not Survived by Gender')
plt.show()
```

```python
sns.barplot(df, x='class', y='fare', hue='sex')
plt.title('Average Fare by passenger class and gender')
plt.show()
```

```python
sns.violinplot(df, x='class', y='age', hue='sex')
plt.title('age vs class')
plt.show()
```

```python
sns.pointplot(df, x='class', y='survived',hue='sex', markers=['o','x'])
plt.title('survival rate by class and gender')
plt.show()
```
